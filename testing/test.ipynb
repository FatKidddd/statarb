```python
import pandas as pd
import numpy as np
import os
```

```python

raw_dfs = []

directory = './alpaca_data/bars/day'
final_df = None
max_n = 0
for i, filename in enumerate(os.listdir(directory)):
  symbol = '.'.join(filename.split('.')[:-1])
  df = pd.read_csv(directory+'/'+filename, index_col='timestamp')
  raw_dfs.append(df)
  
  close = df[['close']]
  symbol_close = close.rename(columns={'close': symbol})

  max_n = max(len(symbol_close), max_n)

  if i == 0:
    final_df = symbol_close
  else:
    final_df = final_df.join(symbol_close, how='outer')
  
  assert max_n == final_df.shape[0]
  # if i == 10:
  #   break
display(max_n)

display(final_df.index)
display(final_df)
```

```python
# generating some test data
timestamp = [1440540000, 1450540000]
df1 = pd.DataFrame({'timestamp': timestamp, 'a': ['val_a', 'val2_a'], 'b': ['val_b', 'val2_b'], 'c': ['val_c', 'val2_c']})
# building a different index
timestamp = timestamp * np.random.randn(abs(1))
df2 = pd.DataFrame({'timestamp': timestamp, 'd': ['val_d', 'val2_d'], 'e': ['val_e', 'val2_e'], 'f': ['val_f', 'val2_f'], 'g': ['val_g', 'val2_g']})
# keeping a value in common with the first index
timestamp = [1440540000, 1450560000]
df3 = pd.DataFrame({'timestamp': timestamp, 'h': ['val_h', 'val2_h'], 'i': ['val_i', 'val2_i']})

# Setting the timestamp as the index
df1.set_index('timestamp', inplace=True)
df2.set_index('timestamp', inplace=True)
df3.set_index('timestamp', inplace=True)

# You can convert timestamps to dates but it's not mandatory I think
df1.index = pd.to_datetime(df1.index, unit='s')
df2.index = pd.to_datetime(df2.index, unit='s')
df3.index = pd.to_datetime(df3.index, unit='s')

display(df1)
display(df2)
display(df3)

# Just perform a join and that's it
result = df1.join(df2, how='outer').join(df3, how='outer')
result
```
